데이터타입이 object임 --> int로 변경 할 필요 있음! <br> 
컬럼 값이 350200 (persons) / '-' 으로 되어있음 <br> 
<br>
전처리 해야할 것 <br>
1. (persons) 제거 <br>
2. '-'제거 <br>
3. 제거 후 남은 공백도 제거 <br>
4. (데이터타입이 object라 공백도 object로 인식) 공백을 null값으로 변경<br>
5. fillna(0)으로 null값을 0 으로 변경 <br>
6. 데이터 타입 float로 변경 <br>
